```python
# Importando as bibliotecas
import tensorflow as tf
import numpy as np
from tensorflow import keras
```

```python
# 1 input, 1 layer, 1 neuronio, rede mais simples
```

```python
model = tf.keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
```

```python
# Devemos especificar: Uma função custo (loss) e um otimizador (optimizer)
model.compile(optimizer='sgd', loss='mean_squared_error')
```

```python
# Criando dados artificias com numpy
# Nosso objetivo é "simular" um conjunto de dados em que 
# a fórmula seja: y = 2x-1

xs = np.array([-1.0,  0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
ys = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype=float)
```

```python
# Treinando a rede neural
model.fit(xs, ys, epochs=500)
```

```python
# Predizendo um valor qualquer
print(model.predict([10.0]))

# Nosso valor esperado seria 19, mas devido a natureza probabilistica 
# das redes neurais por termos poucos dados 
```
